```python
from PIL import Image
import exifread
import pprint

def get_camera_info(image_path):
    # Open the image file
    with open(image_path, 'rb') as f:
        # Open image with PIL
        image = Image.open(f)
        
        # Get EXIF data
        exif_data = image._getexif()
        
        # Check if EXIF data exists
        if exif_data:
            # Parse EXIF data using exifread library
            tags = exifread.process_file(f)
            
            # Extract camera make and model if available
            camera_make = tags.get('Image Make')
            camera_model = tags.get('Image Model')
            
            return camera_make, camera_model, tags
        else:
            return None, None

# Example usage
image_path = '/Users/amtouti/Documents/SFM/3d_sample_data_set_sofa/3D sofas/IMG_1823.jpeg'
camera_make, camera_model, tags = get_camera_info(image_path)
print("Camera Make:", camera_make)
print("Camera Model:", camera_model)
pprint.pprint(tags)
```
